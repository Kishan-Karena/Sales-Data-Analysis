```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales=pd.read_csv("Sales_Data.csv",encoding='unicode_escape')
```

```python
sales.head(15)
```

```python
sales.info()
```

```python
# drop unrelated/blank columns
sales.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
sales.isnull()
```

```python
sales.isnull().sum()
```

```python
sales.shape
```

```python
sales.dropna(inplace=True)
```

```python
sales.shape
```

```python
# change data type
sales['Amount']=sales['Amount'].astype('int')
```

```python
sales['Amount'].dtype
```

```python
sales.columns
```

```python
# rename column
sales.rename(columns={'Marital_Status':'Shaadi'})
```

```python
sales.describe()
```

```python
sales[['Age','Orders','Amount']].describe()
```

```python
# Data Analysis

ax=sns.countplot(x='Gender',data=sales)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender=sales.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

```python
# From above graphs we can see that most of the buyers are females.

ax=sns.countplot(data=sales,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age=sales.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# From above graphs we can see that most of the buyers are of age group between 26-35 y female

#State
sales_state=sales.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# Total amount/sales from top 10 states

sales_state1=sales.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(12)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=sales_state1,x='State',y='Amount')
```

```python
# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh,Maharashtra and Karnataka.

#Marital Status
ax=sns.countplot(data=sales,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state2=sales.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,4)})
sns.barplot(data=sales_state2,x='Marital_Status',y='Amount',hue='Gender')
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchesing power.

# Occupation
ax=sns.countplot(data=sales,x='Occupation')
sns.set(rc={'figure.figsize':(20,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state3=sales.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=sales_state3,x='Occupation',y='Amount')
```

```python
# From above graphs we can see that most of the buyers are working in IT,Healthcare and Aviation sector.

#Product Category
ax=sns.countplot(data=sales,x='Product_Category')
sns.set(rc={'figure.figsize':(20,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state4=sales.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(data=sales_state4,x='Product_Category',y='Amount')
```

```python
# From above Graphs we can see that most of the sold product are from Food,Clothing Electronics Category.

sales_state4=sales.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(22,7)})
sns.barplot(data=sales_state4,x='Product_ID',y='Orders')
```
